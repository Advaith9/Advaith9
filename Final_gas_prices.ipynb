```python
!wget https://raw.githubusercontent.com/sjasthi/python_input_files/main/gasprices.txt

def read_gas_prices(file_name='gasprices.txt'):
    prices_master_list = []
    with open(file_name, 'r') as f:
        for line in f:
            date_price = line.strip().split(':')
            date = date_price[0]
            price = float(date_price[1])
            month = date.split('-')[0]
            day = date.split('-')[1]
            year = date.split('-')[2]
            price_mini_list = [year, month, day, price]
            prices_master_list.append(price_mini_list)
    return prices_master_list


def get_average_price_per_year(prices_master_list):
    start_month = 4
    end_month = 8
    for year in range(1993, 2014):
        total_price = 0
        count = 0
        for price_mini_list in prices_master_list:
            year_str = price_mini_list[0]
            month_str = price_mini_list[1]
            if year_str == str(year) and start_month <= int(month_str) <= end_month:
                total_price += price_mini_list[3]
                count += 1
        if count > 0:
            avg_price = total_price / count
            print(f"Average price in {year}: {avg_price:.3f}")


def get_average_price_per_month(prices_master_list):
    start_year = 1993
    end_year = 2014
    start_month = 4
    end_month = 8

    monthly_prices = {}

    for year in range(start_year, end_year):
        for month in range(1, 13):
            if start_month <= month <= end_month:
                monthly_prices[(year, month)] = []

    for price_mini_list in prices_master_list:
        year = int(price_mini_list[0])
        month = int(price_mini_list[1])
        price = price_mini_list[3]

        if start_month <= month <= end_month and start_year <= year < end_year:
            monthly_prices[(year, month)].append(price)

    for year in range(start_year, end_year):
        for month in range(start_month, end_month+1):
            month_prices = monthly_prices[(year, month)]
            if len(month_prices) > 0:
                avg_price = sum(month_prices) / len(month_prices)
                print(f"Average price in {year}-{month:02d}: {avg_price:.3f}")


# prices_master_list = read_gas_prices()
# get_average_price_per_year(prices_master_list)
# get_average_price_per_month(prices_master_list)

prices_master_list = read_gas_prices()
get_average_price_per_year(prices_master_list)
get_average_price_per_month(prices_master_list)


def get_highest_and_lowest_prices_per_year(prices_master_list):
    year_prices = {}
    for price in prices_master_list:
        year = price[2]
        if year not in year_prices:
            year_prices[year] = []
        year_prices[year].append(price[3])
    for year, prices in year_prices.items():
        highest_price = max(prices)
        lowest_price = min(prices)
        print(f"Highest price in {year}: {highest_price}")
        print(f"Lowest price in {year}: {lowest_price}")

prices_master_list = read_gas_prices()
get_average_price_per_year(prices_master_list)
get_average_price_per_month(prices_master_list)
get_highest_and_lowest_prices_per_year(prices_master_list)

  
```
