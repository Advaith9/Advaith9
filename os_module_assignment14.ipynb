```python
import os
import time
import shutil

def calculate_directory_stats(directory_path):
   #Initialize counters
   num_directories = 0
   num_files = 0
   total_size = 0

    #Walk through the directory and its subdirectories
   for root, dirs, files in os.walk(directory_path):
        num_directories += len(dirs)   #Count directories
        num_files += len(files)        #Count files

        #Calculate total size of files
        for file in files:
                file_path = os.path.join(root, file)
                total_size += os.path.getsize(file_path)

        return num_directories, num_files, total_size

def main():

    print(' 1. Getting the Current Working Directory')

    current_directory = os.getcwd()
    print("Current Directory:", current_directory)

    directory_path = "/Users/sailakshmi/Downloads"

    print("Directory_I'm_scanning:", directory_path)

    print(' 2. Listing Files and Directories in a Directory')

    files_and_dirs = os.listdir(directory_path)
    print("Files and Directories:", files_and_dirs)
    if os.path.exists(directory_path) and os.path.isdir(directory_path):

        num_directories, num_files, total_size = calculate_directory_stats(directory_path)

          #Display the results
        print(f"Number of Directories: {num_directories}")
        print(f"Number of Files: {num_files}")
        print(f"Total Size of Files: {total_size} bytes")

    else:
         print("Invalid directory path. Please provide a valid directory.")

    print(' 3. Creating a Directory')
    new_directory = "/Users/sailakshmi/Documents/New9"

    os.makedirs(new_directory, exist_ok=True)
    os.makedirs(new_directory, exist_ok=False)
    print("Created ", new_directory, "directory")

    print(' 4. Creating a File')
    new_file = "/Users/sailakshmi/Documents/TEST/test_os.txt"
    with open(new_file, 'w') as f:
        f.write("Hello, this is a new file with list of my fav songs!")
    print(new_file, 'is created')

    print(' 5. Renaming a File or Directory')
    old_name = "/Users/sailakshmi/Documents/TEST/test_os.txt"
    new_name = "/Users/sailakshmi/Documents/TEST/new_test_os.txt"
    os.rename(old_name, new_name)
    print(old_name, 'is renamed to', new_name)

    print(' 6. Deleting a File')
    file_to_delete = "/Users/sailakshmi/Documents/Test_Folder/words.txt"
    os.remove(file_to_delete)
    print(file_to_delete, 'is deleted')

    print(' 7. Deleting a Directory')
    directory_to_delete = "/Users/sailakshmi/Documents/Test_Folder"
    os.rmdir(directory_to_delete)
    print(directory_to_delete, 'is deleted')

    print(' 8. Checking if a Path Exists')
    path_to_check = "/Users/sailakshmi/Downloads/OS_Module_test"
    exists = os.path.exists(path_to_check)
    print(f"Path {path_to_check} exists: {exists}")

    print(' 9. Checking if a Path is a File or Directory')
    path_to_check = "/Users/sailakshmi/Downloads/OS_Module_test"
    is_file = os.path.isfile(path_to_check)
    is_directory = os.path.isdir(path_to_check)

    print(f"{path_to_check} is a file: {is_file}")
    print(f"{path_to_check} is a directory: {is_directory}")

    print(' 10. Getting File Information (Size, Last Modified Time, etc.)')
    file_path = "/Users/sailakshmi/Downloads/Dasa_padaamrutham_listofsongs"
    file_info = os.stat(file_path)
    size_in_bytes = file_info.st_size
    last_modified_time = time.ctime(file_info.st_mtime)

    print(f"File Size: {size_in_bytes} bytes")
    print(f"Last Modified Time: {last_modified_time}")

    print(' 11. Copying a File')
    source_file = "/Users/sailakshmi/Downloads/code_v2023-02-14/s3/coffee.jpg"
    destination_file = "/Users/sailakshmi/Downloads/code_v2023-02-14/s3/coffee_bak.jpg"
    shutil.copy(source_file, destination_file)
    print(source_file, 'is copied to', destination_file)


    print(' 12. Copying a Directory')
    source_dir = "/Users/sailakshmi/Downloads/code_v2023-02-14/ssm"
    destination_dir = "/Users/sailakshmi/Documents/ssm_copied"
    shutil.copytree(source_dir, destination_dir)
    print(source_dir, 'is copied to', destination_dir)

    print(' 13. Moving a File')
    source_file = "/Users/sailakshmi/Downloads/2022 Dismissal map.pdf"
    destination_file = "/Users/sailakshmi/Documents/GitHub/PYTHON_ASSIGNMENTS/2022 Dismissal map_1.pdf"
    shutil.move(source_file, destination_file)
    print(source_file, 'is moved to', destination_file)

    print(" 14. Moving a Directory")
    source_dir = "/Users/sailakshmi/Downloads/Math-7-Advanced-YR-Trinh-Raoping--2023-Jan-24_03-43-51-034"
    destination_dir = "/Users/sailakshmi/Documents/"
    shutil.move(source_dir, destination_dir)

    print(" 15.Get the value of the PATH environment variable ")
    env_variable = os.getenv("PATH")
    print("PATH Environment Variable:", env_variable)

    print("Run a system command dir not working on mac -so did ls cmd ")
    command_to_run = "ls"
    result = os.system(command_to_run)
    print("Command Result:", result)

    print("17. Creating a Symbolic Link (Soft Link)")

    source_path = "/Users/sailakshmi/Documents/HTML_resources/Half_Saree.png"
    link_path = "/Users/sailakshmi/Downloads/Half_Saree.png"

    # Check if the source file exists
    if not os.path.exists(source_path):
        print(f"Error: Source file '{source_path}' does not exist.")
    else:
        # Create a symbolic link
        try:
            os.symlink(source_path, link_path)
            print(f"Symbolic link created from '{source_path}' to '{link_path}'.")
        except OSError as e:
            print(f"Error creating symbolic link: {e}")

    print("18. Checking if a Path is an Absolute Path")
    path_to_check = "/Users/sailakshmi/Documents/HTML_resources/Half_Saree.png"
    is_absolute = os.path.isabs(path_to_check)
    print(f"{path_to_check} is an absolute path: {is_absolute}")

    print("19. Joining Paths")
    base_path = "/Users/sailakshmi/Documents/"
    sub_path = "HTML_resources/Half_Saree.png"
    joined_path = os.path.join(base_path, sub_path)
    print("Joined Path:", joined_path)

    print("20. Splitting a Path")
    full_path = "/Users/sailakshmi/Documents/HTML_resources/Half_Saree.png"
    directory, filename = os.path.split(full_path)
    print("Directory:", directory)
    print("Filename:", filename)



if __name__ == "__main__":
    main()
```
