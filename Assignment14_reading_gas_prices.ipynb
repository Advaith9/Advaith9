```python
filename = 'gasprices.txt'

line_number = 1
with open(filename) as gasprices_file_obj:
  for each_line in gasprices_file_obj:
#first splitting the date and price by ":"  
    elements = each_line.split(":")
    date, price = elements
#next splitting the date by the "-"    
    date_parts = date.split("-")
    month, day, year = date_parts
    print("Line number:", line_number)
    print("Month:", month)
    print("Day:", day)
    print("Year:", year)
    print("Price:", price)
    print("")
    line_number += 1
    

```
